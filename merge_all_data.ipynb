```python
import csv
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
funct = pd.read_csv('/content/drive/MyDrive/original_data/patient_data/patient_info.csv')
filenames = list(funct['name'])
print(filenames)
```

```python
old_data=[]
feature_list=[]
with open('/content/drive/MyDrive/original_data/final_data_set/patient_time_npl_v4.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    feature_list=next(csv_reader)
    for row,i in zip(csv_reader,range(0,len(filenames))):
        old_data.append(row)
        for j in range(0,31):
          old_data[i].pop(1)
for i in range(0,31):
  feature_list.pop(1)
print(feature_list)
print(old_data)
```

```python
with open('/content/drive/MyDrive/original_data/final_data_set/MCI_data.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    headers = next(csv_reader)
    headers.pop(0)
    for i in range(0,60):
      feature_list.append(headers[i])
    for row,i in zip(csv_reader,range(0,len(filenames))):
      row.pop(0)
      for j in range(0,60):
        old_data[i].append(row[j])
print(feature_list)
#print(old_data[0:60])
```

```python
with open('/content/drive/MyDrive/original_data/final_data_set/nlp_patient_time_silence.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    headers = next(csv_reader)
    for i in range(len(headers)-2,len(headers)):
      feature_list.append(headers[i])
    for row,i in zip(csv_reader,range(0,len(filenames))):
      for j in range(len(row)-2,len(row)):
        old_data[i].append(row[j])
print(feature_list)
```

```python
with open('/content/drive/MyDrive/original_data/final_data_set/praat_patient_combined.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    headers = next(csv_reader)
    headers.pop(0)
    for i in range(0,len(headers)):
      feature_list.append(headers[i])
    for row,i in zip(csv_reader,range(0,len(filenames))):
      row.pop(0)
      for j in range(0,len(row)):
        old_data[i].append(row[j])
print(feature_list)
```

```python
df = pd.read_csv("/content/drive/MyDrive/original_data/patient_data/patient_info.csv")
result=df['label']
for i in range(0,len(result)):
  if(int(result[i])==1):
    old_data[i].append(1)
  else:
    old_data[i].append(2)
```

```python
feature_list.append('AD_diagnose')
print(feature_list)
```

```python
for i in range(0,len(old_data)):
  #print(old_data[i][feature_list.index('repeat_rate')])
  if(old_data[i][feature_list.index('repeat_rate')]=="inf"):
    old_data[i][feature_list.index('repeat_rate')]=1e9
```

```python
with open('/content/drive/MyDrive/original_data/final_data_set/final_result.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(feature_list)
  for i in range(0,len(old_data)):
    writer.writerow(old_data[i])
```
